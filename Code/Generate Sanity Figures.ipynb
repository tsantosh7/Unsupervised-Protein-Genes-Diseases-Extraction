```python
import pandas as pd
colnames = ['ANN', 'Freq_of_ML_removed', 'Annotation_pipeline_Freq','Percent_removed']
OTAR = pd.read_csv('/nfs/gns/literature/Santosh_Tirunagari/PCSE analysis_OTAR.tsv', sep = '\t', names = colnames, header = None)
```

```python
def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)
```

```python
import string
OTAR_ = OTAR.head(100)
puncs = set(string.punctuation)

ot = {}
for a, x in OTAR_.iterrows():
    if all(x not in puncs for x in x['ANN']) and len(x['ANN'])>2:
        ot[x['ANN']] = x['Freq_of_ML_removed']
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

%matplotlib inline  

wordcloud = WordCloud(background_color="rgba(255, 255, 255, 0)", mode="RGBA", max_font_size=70)
wordcloud.generate_from_frequencies(frequencies=ot)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
